### Web-Based Loader

```python
from langchain_community.document_loaders import WebBaseLoader
web_loader = WebBaseLoader(web_paths=["https://www.bbc.co.uk/programmes/articles/5cRrSfvBF2rV4GHvVbTsVhT/i-think-youre-an-emotional-vampire-five-tips-for-dealing-with-draining-relationships"])
# Spefic Text Extraction from the Web Page
web_loader_specific=WebBaseLoader(web_paths=("https://www.bbc.co.uk/programmes/articles/5cRrSfvBF2rV4GHvVbTsVhT/i-think-youre-an-emotional-vampire-five-tips-for-dealing-with-draining-relationships",),bs_kwargs=dict(parse_only=bs4.SoupStrainer("p")))
```

```python
content=web_loader.load()
content2=web_loader_specific.load()
```
